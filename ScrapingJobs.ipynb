```python
import re

def extract_experience_years(text):
    text = text.lower()

    patterns = [
        r"(?:minimum of )?(\d+)(?: *- *(\d+))? *(?:\+|-)? *years?",
        r"preferred no\. of years \(with relevant degree\) - (\d+)",
        r"(one|two|three|four|five|six|seven|eight|nine|ten) (years?|or more years?)",
        r"(?:year|years)\s*(\d+)"
    ]

    numeric_words = {
        "one": 1,
        "two": 2,
        "three": 3,
        "four": 4,
        "five": 5,
        "six": 6,
        "seven": 7,
        "eight": 8,
        "nine": 9,
        "ten": 10
    }

    max_years = 0
    
    for pattern in patterns:
        matches = re.findall(pattern, text)
        
        for match in matches:
            
            # Pattern 0
            if pattern == patterns[0]:
                if match[1]:
                    years = int(match[1])
                else:
                    years = int(match[0])
            # Pattern 1        
            elif pattern == patterns[1]:
                years = int(match)
            # Pattern 2    
            elif pattern == patterns[2]:
                numeric_word = match[0]
                years = numeric_words.get(numeric_word, 0)
            # Pattern 3   
            elif pattern == patterns[3]:
                years = int(match)
                
            # Check if the years exceed 10
            if years > 10:
                continue
            
            if years > max_years:
                max_years = years

    return max_years
#extract_experience_years(text)
```
